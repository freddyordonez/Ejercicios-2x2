**Simulación de 10,000 lanzamientos de un dado de 6 caras**

```python
# np fundamental para realizar operaciones numéricas eficientes
# Counter clase es muy útil para contar la frecuencia de elementos
# plt me ayudara a la creacion de graficos para ver la recoleccion de datos
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Número de lanzamientos
num_lanzamientos = 10000

# Simular los lanzamientos de dados (valores del 1 al 6)
lanzamientos = np.random.randint(1, 7, size=num_lanzamientos)

# Contar la frecuencia de cada lado
# Counter es un modulo de recoleccion.
frecuencias = Counter(lanzamientos)
```

**Tabla comparativa de Probabilidad Teórica vs. Empírica**

```python
# Probabilidad teórica de que salga cualquier cara en un dado justo
prob_teorica = 1/6

#Imprimen un encabezado formateado para la tabla de resultados
print(f"{'Lado':<6} | {'Frecuencia':<12} | {'P. Empírica':<12} | {'Diferencia':<10}")
print("-" * 50)

# Se inicializan dos listas vacías
# Lados = almacenará los números de las caras del dado (1 al 6)
# probs_emp = guardará las probabilidades empíricas calculadas para cada cara.
lados = []
probs_emp = []

# Inicia un bucle que iterará para cada cara del dado, desde el 1 hasta el 6.

for lado in range(1, 7):

    # Obtiene la frecuencia con la que apareciendo la cara actual.
    frec = frecuencias[lado]

    # Calcula la probabilidad empírica para la cara actual.
    p_emp = frec / num_lanzamientos

    #Calcula la diferencia absoluta entre la probabilidad empírica obtenida y la probabilidad teórica.
    diff = abs(p_emp - prob_teorica)

    # Agrega el número de la cara para el Grafico
    lados.append(lado)
    # Agrega la probabilidad empírica a las listas, esto se usara en el Grafico.
    probs_emp.append(p_emp)

    #Imprime una fila de la tabla con los resultados para la cara actual, mostrando el lado, su frecuencia, su probabilidad empírica
    print(f"  {lado:<4} | {frec:<10} | {p_emp:.4f}      | {diff:.4f}")
```

**Visualizacion**

```python
# Esta línea crea un gráfico de barras con los lados, la probalidad, el color y una asignacion.
plt.bar(lados, probs_emp, color='skyblue', label='Empírica')

# Esta linea de codigo añade una linea horizantal al grafico, que muestra la probalididad teorica aproximadamente 0.1667.
plt.axhline(y=prob_teorica, color='red', linestyle='--', label=f'Teórica ({prob_teorica:.4f})')

# Etiqueta al eje de las X como cara del dado
plt.xlabel('Cara del dado')
# Etiqueta al eje de las y como probalidad
plt.ylabel('Probabilidad')
# Titulo del grafico
plt.title('Simulación de Monte Carlo: 10,000 Lanzamientos')
# Muestra las etiquetas.
plt.legend()
# Cierra y presenta todo lo que va a tener el grafico.
plt.show()
```
